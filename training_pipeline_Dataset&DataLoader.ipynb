```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

```python
y_test
```

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

```python
X_train_tensor.shape,y_train_tensor.shape
```

**Custom dataset**

```python
from torch.utils.data import Dataset,DataLoader
```

```python
class CustomerDataset(Dataset):

  def __init__(self,features,labels):
    self.features=features
    self.labels=labels
  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self, index):
    return self.features[index],self.labels[index]
```

```python
train_dataset=CustomerDataset(X_train_tensor,y_train_tensor)
test_dataset=CustomerDataset(X_test_tensor,y_test_tensor)
```

```python
train_dataset[10]
```

```python
train_loader=DataLoader(train_dataset,batch_size=32,shuffle=True)
test_loader=DataLoader(test_dataset,batch_size=32,shuffle=True)
```

**Create Model**

```python
import torch.nn as nn
```

```python
class SimpleNN(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.linear=nn.Linear(num_features,1)
    self.sigmoid=nn.Sigmoid()

  def forward(self,features):
    out=self.linear(features)
    out=self.sigmoid(out)
    return out

```

```python
learning_rate = 0.1
epochs = 30
```

```python
# create model
model=SimpleNN(X_train_tensor.shape[1])

# optimizer
optimizer=torch.optim.SGD(model.parameters(),lr=learning_rate)

# loss
loss_function=nn.BCELoss()
```

**Training pipeline**

```python
for epoch in range(epochs):
  for b_f,b_l in train_loader:

    # forward padd
    y_pred=model(b_f)

    # loss function
    loss=loss_function(y_pred,b_l.view(-1,1))

    optimizer.zero_grad()

    # backward pas
    loss.backward()

    # paramerters
    optimizer.step()

  print(f"epoch {epoch+1}----> : {loss.item()}")
```

**Model Evaluation**

```python
model.eval()

accuracy_list=[]

with torch.no_grad():
  for b_f,b_l in test_loader:

    y_pred=model(b_f)
    y_pred=(y_pred > 0.6).float()

    batch_accuracy = (y_pred.view(-1) == b_l).float().mean().item()
    accuracy_list.append(batch_accuracy)

overall_accuracy = sum(accuracy_list) / len(accuracy_list)
print(f'Accuracy: {overall_accuracy:.4f}')
```
