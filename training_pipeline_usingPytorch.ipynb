**Train test split**

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns="diagnosis")
x
```

```python
y=df['diagnosis']
y.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**Scaling**

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

Numpy arrays to tensor

```python
import torch
```

```python
X_train_tensor=torch.from_numpy(X_train)
X_train_tensor
```

```python
X_test_tensor=torch.from_numpy(X_test)
X_test_tensor
```

```python
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor.shape,X_test_tensor.shape,y_train_tensor.shape,y_test_tensor.shape
```

**Create Model**




```python
class SimpleNN():

  def __init__(self,X):
    self.weights=torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)
    self.bias=torch.zeros(1,dtype=torch.float64,requires_grad=True)

  def forward_pass(self,X):
    z=torch.matmul(X,self.weights)+self.bias
    y_pred=torch.sigmoid(z)
    return y_pred

  def loss_func(self,y_pred,y):
    epsilon=1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
    return loss



```

```python
model=SimpleNN(X_train_tensor)
```

```python
for epoch in range(0,60):
  y_pred=model.forward_pass(X_train_tensor)

  loss=model.loss_func(y_pred,y_train_tensor)

  loss.backward()

  with torch.no_grad():
    model.weights-=0.1*model.weights.grad
    model.bias-=0.1*model.bias.grad


  model.weights.grad.zero_()
  model.bias.grad.zero_()
  print(f'Epoch: {epoch + 1}, Loss: {loss}')
```

```python
with torch.no_grad():
  y_pred = model.forward_pass(X_test_tensor)
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
